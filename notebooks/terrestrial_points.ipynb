```python
import math
def truncate(number, digits) -> float:
    stepper = 10.0 ** digits
    return math.trunc(stepper * number) / stepper
```

```python
truncate(1234.2858382910,3)
```

```python
fh = open('small.pts')
fh.readline()
fh.readline()
for line in fh:
    fields = line.split(' ')
    out = [truncate(float(i),4) for i in fields]
    print(out)
fh.close()
```
